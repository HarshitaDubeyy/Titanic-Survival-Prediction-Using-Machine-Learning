###**Importing Libraries**





```python
# linear algebra
import numpy as np

# data processing
import pandas as pd

# data visualization
import matplotlib.pyplot as plt

# data visualization
import seaborn as sns

# Importing train_test_split function for splitting dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Importing LogisticRegression class for logistic regression modeling
from sklearn.linear_model import LogisticRegression

# Importing accuracy_score function for evaluating model performance
from sklearn.metrics import accuracy_score
```

###**Data Collection & Processing**

```python
titanic_data = pd.read_csv("/content/train.csv")
```

```python
# printing the first 5 rows of the dataframe
titanic_data.head()
```

```python
# number of rows and columns
titanic_data.shape
```

```python
# getting some informations about the data
titanic_data.info()
```

```python
#check the missing values in each column
titanic_data.isnull().sum()
```

###**Handling the missing values**


```python
#drop the "cabin" column from the dataframe
titanic_data = titanic_data.drop(columns = 'Cabin', axis = 1)
```

```python
#replacing the missing values in "Age" column with mean value
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace =True)
```

```python
#finding the mode value of "Embarked" column
print(titanic_data['Embarked'].mode())
```

```python
print(titanic_data['Embarked'].mode()[0])
```

```python
#replacing the missing values in "Embarked" column with mode value
titanic_data ['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace = True)
```

```python
titanic_data.isnull().sum()
```

###**Encoding the Categorical Columns**

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
# converting categorical columns
titanic_data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}}, inplace = True)
```

```python
titanic_data.head()
```

###**Separating features & Target**

```python
X = titanic_data.drop(columns= ['PassengerId','Name','Ticket','Survived'],axis = 1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

###**Splitting the data into Training data & Test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

###**Model Training**
####*Logistic Regression*

```python
model = LogisticRegression()
```

```python
# training the logistic Regression model with training data
model.fit(X_train, Y_train)
```

###**Model Evaluation**
####*Accuracy Score*

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data :' , training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data :' , test_data_accuracy)
```
